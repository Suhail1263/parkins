IMPORTING THE DEPENDENCIES


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

DATA COLLECTION AND ANALYSIS

```python
#loading the data from csv file to Pandas DataFrame
parkinsons_data=pd.read_csv('parkinsons.data')
```

```python
#printing first 5 rows of dataset
parkinsons_data.head()
```

```python
#number of rows and coloumn in dataset
parkinsons_data.shape
```

```python
parkinsons_data.info()
```

```python
#getting statistical measure about data
parkinsons_data.describe()
```

```python
# distribution of target variable
parkinsons_data['status'].value_counts() 
```

```python
# 1----Person affected by Parkinsons
# 0----Person not affected by Parkinsons
```

```python
parkinsons_data.groupby('status').mean()
```

DATA PRE-PROCESSING


Seprating the features & Target

```python
X= parkinsons_data.drop(columns=['name','status'],axis=1)
```

```python
Y= parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING THE DATA TO TRAINING DATA & TEST DATA

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

DATA STANDARDIZATION

```python
scaler= StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

MODEL TRAINING

SUPPORT VECTOR MACHINE MODEL

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

MODEL EVALUTION

ACCURACY SCORE

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print("Accuracy score of training data : ",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print("Accuracy score of test data : ",test_data_accuracy)
```

BUILD A PREDICTIVE SYSTEM

```python
import pickle
# Writing different model files to file
with open( 'modelForPrediction.sav', 'wb') as f:
    pickle.dump(model,f)
    
with open('standardScalar.sav', 'wb') as f:
    pickle.dump(scaler,f)
```
